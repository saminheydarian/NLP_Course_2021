# Preprocess
The first step of NLP is text preprocessing. Data cleaning is a very crucial step in any machine learning model, but more so for NLP. Without the cleaning process, the dataset is often a cluster of words that the computer doesn‚Äôt understand. Raw data over a properly or improperly formed sentence is not always desirable as it contains lot of unwanted components like null/html/links/url/emoji/stopwords etc. So in this step, this unwanted components are removed for better performance and accuracy.

````python
def clean_html(data):
    cleantext = re.sub('<.*?>', '', data)
    return cleantext
````

````python
def clean_punctuation(data):
    cleantext = re.sub(r'[^\w\s]', '', data)
    return cleantext
````

````python
def lowercase(data):
    lowertext = data.lower()
    return lowertext
````

````python
def word_tokenization(data):
    tokens = nltk.tokenize.word_tokenize(data)
    return tokens
````

````python
def remove_stopwords(tokens):
    stopwords = nltk.corpus.stopwords.words('english')
    cleantext = [i for i in tokens if i not in stopwords]
    return cleantext
````

````python
def lemmatization(tokens):
    wordnet_lemmatizer = WordNetLemmatizer()
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in tokens]
    return lemm_text
````

````python
def clean_data(df):
    # remove html tags
    df['review_html'] = df['review'].apply(clean_html)
    
    # remove punctuations
    df['review_punc'] = df['review_html'].apply(clean_punctuation)
    
    # lower case
    df['review_lower'] = df['review_punc'].apply(lowercase)
    

    # word tokenaztion
    df['review_tokens'] = df['review_lower'].apply(word_tokenization)
    
    
    # remove stop words
    df['review_stopword'] = df['review_tokens'].apply(remove_stopwords)
    
    
    # lemmatization
    df['review_lemm'] = df['review_stopword'].apply(lemmatization)
    
    
    return df
````

````python
df = clean_data(df)
````

````python
# select cleaned feature and class column
df = df[['review_lemm', 'sentiment']]
df.head(10)
````

# Training
Use Naive Beyes algorithm to train a Language Model

````python
class NaiveBayes:
    
    def __init__(self):
        self.pos_prob = 0
        self.neg_prob = 0
        self.pos_word = 0
        self.neg_word = 0
        self.word_set_len = 0
        self.pos_dict = {}
        self.neg_dict = {}
        pass
        
    
    # calculate prior probability for each class
    def prior(self, y_train):
        total_class_count = y_train.shape[0]

        positive = y_train[y_train == 'positive']
        pos_count = positive.shape[0]
        pos_prob = pos_count / total_class_count

        negative = y_train[y_train == 'negative']
        neg_count = negative.shape[0]
        neg_prob = neg_count / total_class_count

        return pos_prob, neg_prob
    
    # claculate number of all words for each class
    def total_word_class(self, train):
        pos_word = 0
        neg_word = 0

        positive = train[train['sentiment'] == 'positive']
        x_pos, y_pos = positive['review_lemm'], positive['sentiment']

        for index, value in x_pos.items():
            pos_word += len(value)

        negative = train[train['sentiment'] == 'negative']
        x_neg, y_neg = negative['review_lemm'], negative['sentiment']

        for index, value in x_neg.items():
            neg_word += len(value)

        return pos_word, neg_word
    
    # calculate vocabulary size
    def word_unique(self, x_train):
        word_list = []
        for index, words in x_train.items():
            for word in words:
                word_list.append(word)

        word_set = set(word_list)
        return len(word_set), len(word_list) 
        
    
    # make dictionary for each class. In each dictionary, we count the number of repetition per word.
    def dictionary(self, train):
        pos_dict = {}
        neg_dict = {}

        positive = train[train['sentiment'] == 'positive']
        x_pos, y_pos = positive['review_lemm'], positive['sentiment']

        for index, words in x_pos.items():
            for word in words:
                if word in pos_dict:
                    pos_dict[word] += 1
                else:
                    pos_dict[word] = 1


        negative = train[train['sentiment'] == 'negative']
        x_neg, y_neg = negative['review_lemm'], negative['sentiment']

        for index, words in x_neg.items():
            for word in words:
                if word in neg_dict:
                    neg_dict[word] += 1
                else:
                    neg_dict[word] = 1

        return pos_dict, neg_dict
    
    # to train naive bayes, we should call all of the above functions. 
    def train(self, train):
        x_train, y_train = train['review_lemm'], train['sentiment']
        self.pos_prob, self.neg_prob = self.prior(y_train)
        self.pos_word, self.neg_word = self.total_word_class(train)
        self.word_set_len, _ = self.word_unique(x_train)
        self.pos_dict, self.neg_dict = self.dictionary(train)
        
    
    # In test we calculate bayes rule for each row in test set.
    # In posterior calculation, Lots of probabilistic multiplication to leads to underflow condition
    # So we apply Log function on posterior term. Log function is monotonic and doesn't have a negative 
    # effect on naive bayes prediction. Multiplication converts to Sum after applying Log.
    def test(self, x_test):
    
        x_test.reset_index(drop=True, inplace=True)
        prediction = np.zeros_like(y_test)


        for index, words in x_test.items():
            pos_likelihood_log = 0
            neg_likelihood_log = 0

            for word in words:
                if word in self.pos_dict:
                    word_prob = (self.pos_dict[word] + 1)/(self.pos_word + self.word_set_len)
                    pos_likelihood_log += np.log(word_prob)
                else:
                    word_prob = (0 + 1)/(self.pos_word + self.word_set_len)
                    pos_likelihood_log += np.log(word_prob)

            pos_prob_log = np.log(self.pos_prob)        
            pos_posterior = pos_likelihood_log + pos_prob_log


            for word in words:
                if word in self.neg_dict:
                    word_prob = (self.neg_dict[word] + 1)/(self.neg_word + self.word_set_len)
                    neg_likelihood_log += np.log(word_prob)
                else:
                    word_prob = (0 + 1)/(self.neg_word + self.word_set_len)
                    neg_likelihood_log += np.log(word_prob)

            neg_prob_log = np.log(self.neg_prob)        
            neg_posterior = neg_likelihood_log + neg_prob_log

            posterior = [neg_posterior, pos_posterior ]
            predicted_label = np.argmax(posterior)
            prediction[index] = predicted_label


        return prediction
        
````

````python
# Instantiate the class
naive_bayes = NaiveBayes()
````

````python
# Call train function on the class
naive_bayes.train(train_set)
````
